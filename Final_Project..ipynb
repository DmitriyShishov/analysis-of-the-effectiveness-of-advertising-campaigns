## Вопрос 2

**Каким образом можно сравнивать несколько моделей регрессии? Какие критерии и подходы при
этом следует применять?**

Для сравнения и оценки моделей регрессии я бы выделил несколько видов:
1) Статистические методы.

Расчет основных метрик: R^2 (коэффициент детерминации, характеризующую обобщающую способность модели), MAE (Средняя абсолютная ошибка не учитывает знак невязки, но зато измеряется в тем же значениях, что и целевая переменная), MSE или RMSE (преимущество их в том, что они учитывают факт того, что невязки модели могут иметь разные знаки)

2) Графические методы

Нанесение предсказанных значений на рельный график целевой переменной может даль лучшее представление о принципах работы модели и указать на слабые места. Применительно к задаче из практической части: градиентный бустинг показывает очень хорошие статистические показатели, но на графике можно увидеть, что он хуже работает в начале даасета, где невозможен расчет лагов или средних показателей за последний период.

3) Представительные методы

Не придумал умного слова, чтобы описать следующие характеристики модели. Для некоторых задач необходима хорошая интерпретируемость модели. Ансамблевые методы могут давать лучшие результаты, но объяснить и понять их работу бывает сложнее, чем условную линейную регрессию. Также отельно бы сказал и про характеристику скорости моделирования. В одних задачах мы можем позволить модели работать несколько часов, в других (например, на биржах) необходимо получать наиболее быстрые результаты.

## Вопрос 3

**Статистика R^2 линейной регрессионной модели составляет 95%. Что это значит? Можно ли
утверждать, что данная модель будет хорошим инструментом прогнозирования зависимой
переменной?**

В изоляции показатель R^2 равный 95% говорит об очень хорошей объяснительной способности. 95% изменений целевой переменной такая модель может объяснить коэффициентами модели. Но в производственных задачах, конечно, необходимо с такой моделью провести еще ряд тестов, таких как:
1) Кросс-валидация

С помощью кросс-валидации можно оценить стабильность модели на разных наборах данных.

2) Проверка на отложенной выборке.

Главный способ проверки любой модели - проверить ее на наборе, который она еще не видела. Большое значение R^2 на тренировочной выборке может свидетельствовать о переобучении. В таком случае на тестовой выборке мы можем получить результат существенно хуже. 

3) Графический анализ.

Также необходимо провести графический анализ результатов. Возможно, модель хорошо работает на больше части данных, но очень плохо на меньшей части данных, где произошло резкое изменение одного из факторов.

4) Дополнительные метрики.

Даже очень хорошее значение R^2 необходимо дополнять прочими метриками. Это также может дать ответ на некоторые особенности модели.

### Работа со столбцом OOH

Первым делом посмотрим на значения, когда OOH не нулевой.

```python
df[df['OOH'] != 0]
```

Можем убедиться, что данный столбец заполняется только в первое число месяца. При этом, несколько месяцев пропущено.

```python
df.groupby(df.Date.dt.to_period("M"))['OOH'].sum().plot(kind = 'bar', title=('Суммарная OTS по наружной рекламе по месяцам'));
```

Оценим график изменения OOH по месяцам. Видим, что наибольшая активность была зарегистрирована в Июле 2014 года, также имеется еще несколько локальных пиков, но никакой конкретной периодики не наблюдается. С натяжкой, но можно сказать про некое повышение активности в летние месяцы.

Так как для прогнозирования нам необходимо иметь значения за каждый день, распределим суммарные значения за месяц равномерно на каждый день.
С помощью оконной функции рассчитаем суммарное значение OOH в каждый месяц и поделим на количество представленных дней в этом месяце.
Далее можем удалить столбец OOH

```python
ooh_per_day = df.groupby(df.Date.dt.to_period("M"))['OOH'].transform(lambda x: x.sum() / x.count())
df['OOH_per_day'] = ooh_per_day
df.drop('OOH', axis=1, inplace=True)
df.head(10)
```

Теперь имея значения активности TV и OOH на каждый день, построим график изменения целевой переменной Y с отмеченными периодами активности рекламы.

```python
plt.figure(figsize=(15, 7))
plt.plot(df['Date'], df['Y'], zorder = 0, label=('График изменения целевой переменной'), c='black')
plt.scatter(df[df['TV'] != 0]['Date'], df[df['TV'] != 0]['Y'], c='r', zorder = 1, s=20, label=('Период активности медиа'))
plt.scatter(df[df['OOH_per_day'] != 0]['Date'], df[df['OOH_per_day'] != 0]['Y'], c='y', zorder = 2, s=5, label=('Период активности наружной рекламы'))
plt.plot(df['Date'], df['Usd_rate'], zorder=3, label=('График изменения курса кубля'), c='b')
plt.legend()
plt.show()
```

### Работа со столбцом Date

Теперь поработаем со столбцом Даты.
Есть предположение, что среднее количество звонков в рабочие и нерабочие дним должны быть разным. И при моделировании этот признак может быть важен.
Из библиотеки holidays извлечем праздники, приуроченные к имеющимся в датасете датам.

```python
import holidays
```

```python
df['hols'] = df['Date'].apply(lambda x: holidays.CountryHoliday('RUS').get(x)).values
df.head()
```

```python
df['hols'].value_counts(dropna=False)
```

Так как нам не так важно, какие это именно праздники, просто пометим праздничные дни единицей, а остальные нулем

```python
df['is_hols'] = df['hols'].astype('bool').astype('int64')
df.drop('hols', axis=1, inplace=True)
df.head()
```

Также извлечем из даты информацию о номере дня недели. Так мы сможем затем получить будни и выходные дни

```python
df['day_of_week'] = df['Date'].dt.dayofweek
df
```

Создадим новый признак is_weekend, где единицей заполним субботу и воскресенье. Остальные дним заполним нулями

```python
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x in [5, 6] else 0)
df.head(10)
```

Теперь на основе столбцов is_hols и is_weekend мы можем получить объединенный столбец с меткой нерабочих (Праздники или выходные) и рабочий дней (остальное).

```python
df['is_non-woking_day'] = df['is_hols'] | df['is_weekend']
df.drop(['is_hols', 'is_weekend'], axis=1, inplace=True)
df.head(10)
```

Построим матрицу корреляции с учетом новых признаков

```python
fig, ax = plt.subplots(figsize=(15, 6))
sns.heatmap(df.corr(), annot=True, vmin=-1, vmax=1, center= 0, cmap= 'coolwarm', linewidths=3, linecolor='black');
```

Как видим, корреляция целевого показателя с новыми признаками действительно имеется. Она отрицательная, что говорит нам о том, что количество звонков увеличивается в будние дни недели и в рабочие (не праздники и не выходные)

Построим сводную таблицу на основе рабочих и нерабочих дней и убедимся, что значения признаков для них разнятся

```python
df.groupby('is_non-woking_day').mean()
```

Также построим графики средних значений признаков в рабочие и нерабочие дни.

### Генерация новых признаков

Теперь создадим еще несколько дополнительных признаков для улучшения моделирования

```python
df2 = df.copy()
```

Так как мы работаем с временными рядами, где каждое следующее положение может зависеть от предыдущего, использование лагов является допустимым. Поэтому рассчитаем 3 лага для основных признаков. 
Также рассчитаем для этих признаков среднее значение в скользящем окне, что поможет учесть двухнедельную и месячную тенденции.

```python
for col in ['Y', 'TV', 'OOH_per_day', 'Seasonal_Sales', 'Usd_rate']:
    for lag in range(1, 4):
        df2[f'{col}_lag_{lag}'] = df2[col].shift(lag)
    df2[f'Rolling_Average_{col}_15'] = df2[col].rolling(window=15).mean()
    df2[f'Rolling_Average_{col}_30'] = df2[col].rolling(window=30).mean()
```

```python
df2.head()
```

Заполним образовавшиеся после расчета лагов пустые значения в начале датасета с помощью линейной интерполяции

```python
df2.interpolate(method='linear', limit_direction='both', inplace=True)
df2.head()
```

Также для дополнительного учета изменений целевой переменной в рамках одного месяца, рассчитаем отношение текущего значения Y к среднему по месяцу.

```python
df2['Y_to_month'] = df2.groupby([df2.year, df2.month])['Y'].transform(lambda x: x / x.mean())
df2.head(10)
```

Также для дополнительго учета общего влияния основных признаков, сгенерируем новые столбцы, описывающие их произведение

```python
df2['TV_Season_Sales_interaction'] = df2['TV'] * df2['Seasonal_Sales']
df2['TV_OOH_interaction'] = df2['TV'] * df2['OOH_per_day']
df2['OOH_Season_Sales_interaction'] = df2['OOH_per_day'] * df2['Seasonal_Sales']
df2.head(10)
```

## Моделирование

```python
x_train = df_train.drop('Y', axis=1)
y_train = df_train['Y']

x_test = df_test.drop('Y', axis=1)
y_test = df_test['Y']
```

Далее для нескольких моделей регрессии оценим среднее значение коэффициента детерминации (R^2) по результатам кросвалидации для тренировочного датасета.
Затем используем модель для предсказания целевого признака на отложенной выборке. Сохраним основные метрики каждой модели в отдельный датафрейм

```python
models = [LinearRegression(), RandomForestRegressor(), KernelRidge(), ElasticNet(), GradientBoostingRegressor()]
model_metrics = []
for model in models:
    model.fit(x_train, y_train)
    
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(model, x_train, y_train, cv=kfold, scoring='r2')
    train_r2 = cv_scores.mean()
    train_std = cv_scores.std()

    y_test_pred = model.predict(x_test)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    model_metrics.append({
        'model': (f'{model}'),
        'Среднее R^2 на кросс-валидации': round(train_r2, 3),
        'Стандартное отклонение на кросс-валидации': round(train_std, 3),
        'MSE на отложенной выборке': round(test_mse, 3),
        'MAE на отложенной выборке': round(test_mae, 3),
        'R^2 на отложенной выборке': round(test_r2, 3)
    })

result_df = pd.DataFrame(model_metrics)
```

```python
result_df
```

Средние значения R^2 почти на каждой модели показывают очень хорошие результаты что на тренировочной, что на тестовой выборке.

### Работа с моделью ElasticNet

ElasticNet является надежной моделью, которая включает в себя возможность комбинировать L1 и L2 регуляризацию. Протестируем параметры alpha и l1_ratio

```python
param_grid = {
    'alpha': [0.1, 0.5, 1],
    'l1_ratio': [0, 0.25, 0.5, 0.75, 1]
}

kfold = KFold(n_splits=5, shuffle=True, random_state=42)
grid_search = GridSearchCV(ElasticNet(max_iter=1000), param_grid, cv=kfold, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(x_train, y_train)

print("Лучший параметр alpha:", grid_search.best_params_['alpha'])
print("Лучшее значение l1_ratio:", grid_search.best_params_['l1_ratio'])
```

Используем наилучшие параметры для финаального моделирования и сохраним метрики в датасет

```python
elnet_model = ElasticNet(alpha=0.1, l1_ratio=1)
elnet_model.fit(x_train, y_train)
    
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(elnet_model, x_train, y_train, cv=kfold, scoring='r2')
train_r2 = cv_scores.mean()
train_std = cv_scores.std()
print(f'Среднее R^2 на кросс-валидации: {round(train_r2, 3)}')
print(f'Стандартное отклонение на кросс-валидации: {round(train_std, 3)}')

y_test_pred_elnet = elnet_model.predict(x_test)

test_mse = mean_squared_error(y_test, y_test_pred_elnet)
test_mae = mean_absolute_error(y_test, y_test_pred_elnet)
test_r2 = r2_score(y_test, y_test_pred_elnet)

print(f'MSE на отложенной выборке: {round(test_mse, 3)}')
print(f'MAE на отложенной выборке: {round(test_mae, 3)}')
print(f'R^2 на отложенной выборке: {round(test_r2, 3)}')

model_metrics.append({
        'model': ('ElasticNet_tuned'),
        'Среднее R^2 на кросс-валидации': round(train_r2, 3),
        'Стандартное отклонение на кросс-валидации': round(train_std, 3),
        'MSE на отложенной выборке': round(test_mse, 3),
        'MAE на отложенной выборке': round(test_mae, 3),
        'R^2 на отложенной выборке': round(test_r2, 3)
    })
```

Запишем предсказания модели в датасет и построим график реальных значений Y и предсказанных на тестовой выборке

```python
df_test['Y_pred_elnet'] = y_test_pred_elnet.astype(int)
fig, ax = plt.subplots(figsize=(15, 6))
plt.plot(df.index, df['Y'], zorder = 0, label = 'График изменения целевой переменной')
plt.scatter(df_test.index, df_test['Y'], s=30, c='y', zorder=1, label = 'Реальные значения Y на тестовой выборке')
plt.scatter(df_test.index, df_test['Y_pred_elnet'], s = 20, c='r', zorder=2, marker='*', label = 'Предсказанные значения Y на тестовой выборке')
plt.title('Результаты предсказания наилучшей модели ElasticNet')
plt.legend()
plt.show()
```

GradientBoostingRegressor - ансамблевый метод, который основан на построение нескольких моделей (деревья решений), каждая из который пытается учесть и сминимизировать ошибки предыдущей.

```python
param_grid = {
    'learning_rate': [0.1, 0.5, 1],
    'max_depth': [None, 5, 10],
    'subsample': [0.1, 0.5, 0.9],
    'n_estimators': [100, 200, 300]
}

kfold = KFold(n_splits=5, shuffle=True, random_state=42)
grid_search = GridSearchCV(GradientBoostingRegressor(), param_grid, cv=kfold, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(x_train, y_train)

print("Лучший параметр learning_rate:", grid_search.best_params_['learning_rate'])
print("Лучшее значение max_depth:", grid_search.best_params_['max_depth'])
print("Лучший параметр subsample:", grid_search.best_params_['subsample'])
print("Лучшее значение n_estimators:", grid_search.best_params_['n_estimators'])
```

```python
gbr_model = GradientBoostingRegressor(learning_rate=0.1, max_depth=5, subsample=0.5, n_estimators=100)
gbr_model.fit(x_train, y_train)
    
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(gbr_model, x_train, y_train, cv=kfold, scoring='r2')
train_r2 = cv_scores.mean()
train_std = cv_scores.std()
print(f'Среднее R^2 на кросс-валидации: {round(train_r2, 3)}')
print(f'Стандартное отклонение на кросс-валидации: {round(train_std, 3)}')

y_test_pred_gbr = gbr_model.predict(x_test)
test_mse = mean_squared_error(y_test, y_test_pred_gbr)
test_mae = mean_absolute_error(y_test, y_test_pred_gbr)
test_r2 = r2_score(y_test, y_test_pred_gbr)

print(f'MSE на отложенной выборке: {round(test_mse, 3)}')
print(f'MAE на отложенной выборке: {round(test_mae, 3)}')
print(f'R^2 на отложенной выборке: {round(test_r2, 3)}')

model_metrics.append({
        'model': ('GradientBoostingRegressor_tuned'),
        'Среднее R^2 на кросс-валидации': round(train_r2, 3),
        'Стандартное отклонение на кросс-валидации': round(train_std, 3),
        'MSE на отложенной выборке': round(test_mse, 3),
        'MAE на отложенной выборке': round(test_mae, 3),
        'R^2 на отложенной выборке': round(test_r2, 3)
    })
```

### Работа с моделью RandomForestRegressor

RandomForestRegressor - еще один ансамлевый метод, который основан на сборе и осреднении результатов нескольких деревьев решений.Каждое дерево работает по принципу разделения пространства данных по наиболее отличающемуся значению признака.

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

kfold = KFold(n_splits=5, shuffle=True, random_state=42)
grid_search = GridSearchCV(RandomForestRegressor(), param_grid, cv=kfold, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(x_train, y_train)

print("Лучший параметр n_estimators:", grid_search.best_params_['n_estimators'])
print("Лучшее значение max_depth:", grid_search.best_params_['max_depth'])
print("Лучший параметр min_samples_split:", grid_search.best_params_['min_samples_split'])
print("Лучшее значение min_samples_leaf:", grid_search.best_params_['min_samples_leaf'])
```

```python
rf_model = RandomForestRegressor(n_estimators=300, max_depth=None, min_samples_split=5, min_samples_leaf=1)
rf_model.fit(x_train, y_train)
    
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(rf_model, x_train, y_train, cv=kfold, scoring='r2')
train_r2 = cv_scores.mean()
train_std = cv_scores.std()
print(f'Среднее R^2 на кросс-валидации: {round(train_r2, 3)}')
print(f'Стандартное отклонение на кросс-валидации: {round(train_std, 3)}')

y_test_pred_rf = rf_model.predict(x_test)
test_mse = mean_squared_error(y_test, y_test_pred_rf)
test_mae = mean_absolute_error(y_test, y_test_pred_rf)
test_r2 = r2_score(y_test, y_test_pred_rf)

print(f'MSE на отложенной выборке: {round(test_mse, 3)}')
print(f'MAE на отложенной выборке: {round(test_mae, 3)}')
print(f'R^2 на отложенной выборке: {round(test_r2, 3)}')

model_metrics.append({
        'model': ('RandomForestRegressor_tuned'),
        'Среднее R^2 на кросс-валидации': round(train_r2, 3),
        'Стандартное отклонение на кросс-валидации': round(train_std, 3),
        'MSE на отложенной выборке': round(test_mse, 3),
        'MAE на отложенной выборке': round(test_mae, 3),
        'R^2 на отложенной выборке': round(test_r2, 3)
    })
```
